```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv("normalized_sc_2000vg_baron.csv")
df
```

```python
df = df.drop('Unnamed: 0', axis=1)
df
```

```python
label = pd.read_csv("Labels.csv")
label 
```

```python
labels = label.x
labels
```

```python
labels.value_counts()
```

```python
X_train, X_test, y_train, y_test = train_test_split(df, labels, stratify=labels,test_size=0.2, random_state=42)
```

```python
X_train.to_csv("EnProCell_tool/train_set.csv")
```

```python
y_train.to_csv("EnProCell_tool/train_label.csv")
```

```python
X_test.to_csv("EnProCell_tool/test_set.csv")
```

```python
y_test.to_csv("EnProCell_tool/test_label.csv")
```
